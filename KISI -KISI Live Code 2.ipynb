#####3. Berapa banyak total peminjaman sepeda pada tahun 2016 dan 2017?

````python
# working area
query = """
SELECT COUNT(*) AS total_peminjaman
FROM `ftds-hacktiv8-project.practice.bike_rent`
WHERE EXTRACT(YEAR FROM TIMESTAMP(start_time)) BETWEEN 2016 AND 2017
"""
df_1 = client.query(query).to_dataframe()
df_1


````

#####4. Dimana station yang paling banyak dikunjungi sebagai titik keberangkatan pada tahun 2016?

````python
# working area
query = """
SELECT start_station_name, COUNT(*) AS total_departures
FROM `ftds-hacktiv8-project.practice.bike_rent`
WHERE EXTRACT(YEAR FROM start_time) = 2016
GROUP BY start_station_name
ORDER BY total_departures DESC
LIMIT 1;
"""

df_1 = client.query(query).to_dataframe()
df_1




````

#####5. Berapa nilai minimum, maximum, dan rata-rata dari durasi perjalanan pada keseluruhan data?

````python
# working area
query = """
SELECT
    MIN(duration_minutes) AS min_duration,
    MAX(duration_minutes) AS max_duration,
    AVG(duration_minutes) AS avg_duration
FROM `ftds-hacktiv8-project.practice.bike_rent`;
"""
df_1 = client.query(query).to_dataframe()
df_1
````

#####6. Berapa rata-rata durasi perjalanan dari subscriber_type `Student Membership` dan `Local31`?

````python
# working area
query = """
SELECT
    subscriber_type,
    AVG(duration_minutes) AS avg_duration
FROM `ftds-hacktiv8-project.practice.bike_rent`
WHERE subscriber_type IN ('Student_Membership', 'Local31')
GROUP BY subscriber_type;
"""
df_1 = client.query(query).to_dataframe()
df_1




````

#####2. Buat kolom baru bernama `quarter`.
  
  Lakukan ekstraksi waktu (quarter dalam tahun) berdasarkan kolom `start_time` dengan ketentuan berikut:

<div align="center">

  |Keterangan Waktu|Bulan|
  |--|--|
  |Q1|Januari, Februari, Maret|
  |Q2|April, Mei, Juni|
  |Q3|Juli, Agustus, September|
  |Q4|Oktober, November, Desember|

</div>

>**Notes:**
>- Dalam menjawab soal Pandas no. 2, **tidak boleh** menggunakan ekstraksi waktu `Series.dt.quarter`.
- Lakukan `looping` dan `conditional` dengan memanfaatkan ekstraksi bulan `Series.dt.month` terhadap kolom `start_time`, pastikan tipe datanya ialah `datetime` agar dapat dilakukan ekstraksi.
- Simpan hasil looping ke dalam suatu variabel, dan buat kolom baru berisikan variabel sebelumnya.

````python
# working area
import pandas as pd

# Contoh DataFrame
data = {
    'start_time': ['2024-01-15 08:00:00', '2024-05-20 10:30:00', '2024-08-05 14:00:00', '2024-11-30 09:00:00']
}
df = pd.DataFrame(data)

# Mengonversi kolom 'start_time' menjadi datetime
df['start_time'] = pd.to_datetime(df['start_time'])

# Ekstraksi bulan
months = df['start_time'].dt.month

# Menentukan kuartal berdasarkan bulan
def get_quarter(month):
    if month in [1, 2, 3]:
        return 'Q1'
    elif month in [4, 5, 6]:
        return 'Q2'
    elif month in [7, 8, 9]:
        return 'Q3'
    elif month in [10, 11, 12]:
        return 'Q4'

# Terapkan fungsi ke setiap bulan dan buat kolom baru 'quarter'
df['quarter'] = months.apply(get_quarter)

# Tampilkan hasil
print(df)

````

#####3. Tampilkan data perjalanan dari tipe member `Local31` dengan durasi perjalanan diatas 1 jam.
*(Jika tidak ada entry yang memenuhi kondisi ini tidak apa, tuliskan kode pandasnya dengan lengkap saja)*

````python
# working area
import pandas as pd

# Contoh DataFrame
data = {
    'member_type': ['Local31', 'Local31', 'Local31'],
    'trip_duration': [90, 120, 75]  # durasi dalam menit
}
df = pd.DataFrame(data)

# Mengonversi kolom 'trip_duration' dari menit ke jam
df['trip_duration_hours'] = df['trip_duration'] / 60

# Filter data berdasarkan tipe member dan durasi perjalanan
filtered_df = df[(df['member_type'] == 'Local31') & (df['trip_duration_hours'] > 1)]

# Tampilkan hasil
print(filtered_df)
````

#####4. Berapa rata-rata durasi perjalanan dari tipe pelanggan `Walk Up` dengan tipe sepeda `classic` pada bulan `Maret` dari data yang diambil?
*(Jika tidak ada entry yang memenuhi kondisi ini tidak apa, tuliskan kode pandasnya dengan lengkap saja)*

````python
# working area
import pandas as pd

# Contoh DataFrame
data = {
    'start_time': ['2024-03-01 08:00:00', '2024-03-15 10:30:00', '2024-02-05 14:00:00', '2024-03-25 09:00:00'],
    'member_type': ['Walk Up', 'Walk Up', 'Walk Up', 'Walk Up'],
    'bike_type': ['classic', 'classic', 'mountain', 'classic'],
    'trip_duration': [90, 120, 60, 80]  # durasi dalam menit
}
df = pd.DataFrame(data)

# Mengonversi kolom 'start_time' menjadi datetime
df['start_time'] = pd.to_datetime(df['start_time'])

# Ekstraksi bulan dari kolom 'start_time'
df['month'] = df['start_time'].dt.month

# Filter data berdasarkan tipe pelanggan, tipe sepeda, dan bulan Maret
filtered_df = df[(df['member_type'] == 'Walk Up') &
                 (df['bike_type'] == 'classic') &
                 (df['month'] == 3)]

# Hitung rata-rata durasi perjalanan dalam menit
avg_duration = filtered_df['trip_duration'].mean()

# Tampilkan hasil
print(f"Rata-rata durasi perjalanan: {avg_duration} menit")
````
